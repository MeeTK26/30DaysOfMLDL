```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("boston.csv")
df.head()
```

```python
y=df.MEDV.values
```

```python
data=np.array(df.drop("MEDV",axis=1))
data
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(data)
X=scaler.transform(data)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=45)
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
print("By use of simple linear regression")
print("r2 score is ",r2_score(y_test,y_pred))
print("mean square error is ",mean_squared_error(y_test,y_pred))
```

```python
from sklearn.linear_model import Ridge
model=Ridge(alpha=0.01)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
print("By use of ridge regression")
print("r2 score is ",r2_score(y_test,y_pred))
print("mean square error is ",mean_squared_error(y_test,y_pred))
```

```python
from sklearn.linear_model import Lasso
model=Lasso(alpha=0.01)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
print("By use of lasso regression")
print("r2 score is ",r2_score(y_test,y_pred))
print("mean square error is ",mean_squared_error(y_test,y_pred))
```

```python


import matplotlib.pyplot as plt


plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Linear Regression: Actual vs. Predicted")
plt.show()


plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Ridge Regression: Actual vs. Predicted")
plt.show()


plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Lasso Regression: Actual vs. Predicted")
plt.show()


r2_scores = [r2_score(y_test, y_pred), r2_score(y_test, y_pred),r2_score(y_test, y_pred)]
models = ['Linear Regression', 'Ridge Regression', 'Lasso Regression']

plt.figure(figsize=(8, 6))
plt.bar(models, r2_scores)
plt.ylabel('R2 Score')
plt.title('Comparison of R2 Scores for Different Models')
plt.show()

```
