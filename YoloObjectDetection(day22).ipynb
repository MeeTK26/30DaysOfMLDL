```python
from ultralytics import YOLO
import cv2
```

```python
model = YOLO("yolov8n.pt")
```

```python
image_path = "coco_dataset/val2017/000000039769.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
results = model(image)
for result in results:
    result.show()
```
